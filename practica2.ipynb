DataFrames de pandas
------------------------------------
un dataframe es una estructura de datos bidimencional,como una matriz o tabla con filas.
-----------------------------
 ejemplo: crea un dataframe basico:
----------------------------------


```python
import pandas as pd
datos={
    "calorias": [420,380,390],
    "duracion": [50,40,45]
}
#cargar datos dentro del DataFrame
dataframe=pd.DataFrame(datos)
print(dataframe)
```

Localizar filas
Como puede ver el resultado anterior el DataFrame es como una tabla con filas y columnas. Pandas usa el atributo "loc" para devolver una o mas filas especificas

Ejemplo:Devuelve la fila 0

```python
import pandas as pd
datos = {
    "calorias":[420,380,390],
    "duracion":[50,40,45]
}
#cargar datos de DATAFRAME
df = pd.DataFrame(datos)
print(df.loc[0])
```

Devuelve las filas 0 y 1:

```python
import pandas as pd
datos = {
    "calorias":[420,380,390],
    "duracion":[50,40,45]
}
#cargar datos de DATAFRAME
df = pd.DataFrame(datos)
print(df.loc[[0,1]])
```

Indices nombrados
-----------------------
con el argumento "index", puedes nombrar tus propios indices o etiquetas.
---------------------------------
Ejemplo: Agrega una lista de nombres para darle etiqueta a cada fila.
------------------------------

```python
import pandas as pd
datos = {
    "calorias":[420,380,390],
    "duracion":[50,40,45]
}
df = pd.DataFrame(datos, index = ["dia1","dia2","dia3"])
print(df)
```

localizar indices por el nombre o etiqueta utilica el atributo "loc" para devolver la(s) fila(s) especificas ejemplo 2: devuelve fila dia 2

```python
import pandas as pd
data={
    "calorias":[420,380,390],
    "duracion":[50,40,45]
}
df=pd.DataFrame(data,index=["dia 1","dia 2","dia 3"])
print(df.loc["dia 2"])
```
